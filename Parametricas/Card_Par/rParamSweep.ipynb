```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import pandas as pd
```

```python
df16 = pd.read_excel('rParamSweep.xlsx', 'r=8.8', decimal=',')
df16['Frequency / GHz'] = round(df16['Frequency / GHz'],2)*10**9
df17 = pd.read_excel('rParamSweep.xlsx', 'r=9.9', decimal=',')
df17['Frequency / GHz'] = round(df17['Frequency / GHz'],2)*10**9
df18 = pd.read_excel('rParamSweep.xlsx', 'r=11', decimal=',')
df18['Frequency / GHz'] = round(df18['Frequency / GHz'],2)*10**9
df19 = pd.read_excel('rParamSweep.xlsx', 'r=12.1', decimal=',')
df19['Frequency / GHz'] = round(df19['Frequency / GHz'],2)*10**9
df20 = pd.read_excel('rParamSweep.xlsx', 'r=13.2', decimal=',')
df20['Frequency / GHz'] = round(df20['Frequency / GHz'],2)*10**9
```

```python
# figure(figsize=(30, 24), dpi=300)
fig, ax = plt.subplots(layout = 'constrained',figsize=(6.4*2,4.8*2))
ax.plot(df16['Frequency / GHz'], df16['S1,1'], label ='R = 8.8 mm', lw = 0.8)
ax.plot(df17['Frequency / GHz'], df17['S1,1'], label ='R = 9.9 mm', lw = 0.8)
ax.plot(df18['Frequency / GHz'], df18['S1,1'], label ='R = 11.0 mm', lw = 0.8)
ax.plot(df19['Frequency / GHz'], df19['S1,1'], label ='R = 12.1 mm', lw = 0.8)
ax.plot(df20['Frequency / GHz'], df20['S1,1'], label ='R = 13.2 mm', lw = 0.8)

plt.axhline(y=-10, ls=':', c='r') #Horizontal Line
plt.ticklabel_format(axis='x', style='sci', scilimits=(9,9))
plt.xticks(range(0, 28*10**9, 2*10**9)) #x scale
ax.set_xlabel('Frequency [GHz]')  # Add an x-label to the axes.
ax.set_ylabel('S(1,1) [dB]')  # Add a y-label to the axes.
ax.set_title('S-Parameters [Magnitude in dB]')  # Add a title to the axes.
ax.set(xlim=(0, max(df16['Frequency / GHz'])), ylim=(-40, 0))
ax.grid()
ax.legend()
```
