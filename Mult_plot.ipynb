```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('Data Analysis\Data.xlsx', '2nd_inset_analysis', decimal=',')
df.rename(columns = {'dB(S(1,1))[0, ::]':'dB'}, inplace = True)
df
```

```python
newdf=df.loc[(df.lins== 0.5)]
newdf
```

```python
df05 = df.loc[df.lins == 0.5]
df05
```

```python
highest_freq = 1.5e+10

df05 = df.loc[df.lins == 0.5]
df15 = df.loc[df.lins == 1.5]
df25 = df.loc[df.lins == 2.5]
df35 = df.loc[df.lins == 3.5]
df45 = df.loc[df.lins == 4.5]
df55 = df.loc[df.lins == 5.5]
df65 = df.loc[df.lins == 6.5]
df75 = df.loc[df.lins == 7.5]
df85 = df.loc[df.lins == 8.5]
df95 = df.loc[df.lins == 9.5]
df105 = df.loc[df.lins == 10.5]
df115 = df.loc[df.lins == 11.5]

fig, ax = plt.subplots( layout = 'constrained')

ax.plot(df05['freq'],df05['dB'], label ='0.5')
ax.plot(df15['freq'],df15['dB'], label ='1.5')
ax.plot(df25['freq'],df25['dB'], label ='2.5')
ax.plot(df35['freq'],df35['dB'], label ='3.5')
ax.plot(df45['freq'],df45['dB'], label ='4.5')
ax.plot(df45['freq'],df55['dB'], label ='5.5')
ax.plot(df65['freq'],df65['dB'], label ='6.5')
ax.plot(df75['freq'],df75['dB'], label ='7.5')
ax.plot(df85['freq'],df85['dB'], label ='8.5')
ax.plot(df95['freq'],df95['dB'], label ='9.5')
ax.plot(df105['freq'],df105['dB'], label ='10.5')
ax.plot(df115['freq'],df115['dB'], label ='11.5')

plt.axhline(y=-10, ls=':', c='r') #Horizontal line
ax.set_xlabel('Frequency')  # Add an x-label to the axes.
ax.set_ylabel('S(1,1) [dB]')  # Add a y-label to the axes.
ax.set(xlim=(0, highest_freq), ylim=(-20, 0)) #Set limits
ax.set_title('S11_Analysis')  # Add a title to the axes.
ax.grid()
ax.legend()

plt.show
```
