```python
import pandas as pd
import matplotlib.pyplot as plt

xl = pd.ExcelFile('Parametricas\\Koch_Par\\W\\W.xlsx')

sheets = xl.sheet_names

df1 = xl.parse(sheets[0])
df2 = xl.parse(sheets[1])
df3 = xl.parse(sheets[2])
df4 = xl.parse(sheets[3])
df5 = xl.parse(sheets[4])
df6 = xl.parse(sheets[5])

fig, ax = plt.subplots(layout = 'constrained', figsize=(6.4*2,4.8*2))
ax.plot(df6['Frequency / GHz']*10**9, df6['S1,1'], ls='-',label ="W = "+sheets[0]+" mm", linewidth = 1.5)
ax.plot(df2['Frequency / GHz']*10**9, df2['S1,1'], ls='-',label ="W = "+sheets[1]+" mm", linewidth = 0.5)
ax.plot(df3['Frequency / GHz']*10**9, df3['S1,1'], ls='-',label ="W = "+sheets[2]+" mm", linewidth = 0.5)
ax.plot(df4['Frequency / GHz']*10**9, df4['S1,1'], ls='-',label ="W = "+sheets[3]+" mm", linewidth = 0.5)
ax.plot(df5['Frequency / GHz']*10**9, df5['S1,1'], ls='-',label ="W = "+sheets[4]+" mm", linewidth = 0.5)


plt.axhline(y=-10, ls=':', c='r') #Horizontal Line
plt.ticklabel_format(axis='x', style='sci', scilimits=(9,9))
plt.xticks(range(0, 28*10**9, 1*10**9)) #x scale
plt.yticks(range(-30, 2, 2))
ax.set_xlabel('Frequencia [GHz]')  # Add an x-label to the axes.
ax.set_ylabel('S(1,1) [dB]')  # Add a y-label to the axes.
# ax.set_title('S11_Analysis')  # Add a title to the axes.
ax.set(xlim=(0, max(df1['Frequency / GHz']*10**9)), ylim=(-30, 0))
ax.grid()
ax.legend()

```
